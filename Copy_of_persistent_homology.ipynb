```python
from matplotlib import pyplot as plt
plt.ion()
feature = spit # inlet # cape #
```

```python
plt.scatter(*feature.T, s=1)
plt.gca().set_aspect('equal')
```

```python
#line = feature.copy()#feature[:,0],np.zeros_like(feature[:,1])
#line[:,1] = 0
line = np.array([np.linspace(min(feature[:,0]),max(feature[:,0]),len(feature[:,0])),np.zeros_like(feature[:,1])])
plt.scatter(*line, s=1)
plt.gcf().set_size_inches(6.5, 1)
```

```python
import ripser
feature_dgm = ripser.ripser(feature)['dgms'][0]
line_dgm = ripser.ripser(line.T)['dgms'][0]
persim.plot_diagrams([feature_dgm, line_dgm] , labels=['feature', 'line'])
```

```python
%matplotlib inline
dist, matching = persim.bottleneck(feature_dgm, line_dgm, matching=True)
print(dist)
plt.clf()
persim.bottleneck_matching(feature_dgm, line_dgm, matching, labels=['feature', 'line'])
plt.show()
```
